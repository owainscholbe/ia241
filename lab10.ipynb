```python
import xlrd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
book = xlrd.open_workbook('house_price.xls')
sheet = book.sheet_by_name('price')
num_record = sheet.nrows

price_list = []
bathroom_list = []
type_list = []
lot_list = []
year_list = []
area_list = []

for i in range(num_record):
    row = sheet.row_values(i)
    price, bathroom, house_type, lot_size, year, area = row
    if i != 0:
        #print(price, bathroom, house_type, lot_size, year, area)
        price_list.append(price)
        bathroom_list.append(bathroom)
        type_list.append(house_type)
        lot_list.append(lot_size)
        year_list.append(year)
        area_list.append(area)
        
print(price_list)
print(bathroom_list)
print(type_list)
print(lot_size)
print(year_list)
print(area_list)
```

#### 2)

```python
plt.plot(year_list, price_list)
plt.ylabel('House Price')
plt.xlabel('Year Built')
plt.show()
```

#### 3)

```python
plt.hist(price_list)
plt.show()
```

#### 4)

```python
plt.scatter(area_list, price_list, s = bathroom_list)
plt.xlabel('House Area')
plt.ylabel('House Price')
plt.show()
```
